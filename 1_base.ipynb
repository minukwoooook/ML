# 붓꽃 품종 나누기
- 분류(classification문제에 해당)
- 출력될 수 있는 값(붓꽃의 종류)들을 class라고 함

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
iris_dataset = load_iris()
```

```python
#X:data y:label
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
#train_test_split 함수 사용 시 75%를 train set으로, 25%를 test set으로 설정
print("X_train size:",X_train.shape)
print("X_test size:",X_test.shape)
```

# KNN(K-Nearest Neighbors)
- 훈련 데이터 셋으로 모델을 만들고 싶을 땐 fit method이용

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred = knn.predict(X_test)
print("test set 예측 값:",y_pred)
```

```python
print("test set 예측 정확도:{:.2f}".format(np.mean(y_pred==y_test)))
print("test set 정확도:{:.2f}".format(knn.score(X_test,y_test)))
```
